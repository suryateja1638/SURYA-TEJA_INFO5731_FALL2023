Question 2 (10 points): Write python code to extract these features you discussed above. You can collect a few sample text data for the feature extraction.

```python
# You code here (Please add comments in the code):
from six.moves import urllib
import pandas as pd

t1 = []
url = "https://raw.githubusercontent.com/srujan-bandla/srujan_INFO5731_Fall2022/main/In_class_exercise/sampledata.txt"
tf = urllib.request.urlopen(url)  #This function is used to decode the url and get the data from it it is a predefined function

for i in tf:
    dec_data = i.decode("utf-8").replace('\r\n', '')    #this statement is used to decode the data present in the url and convert it into standard format
    if dec_data:
        t1.append(dec_data)       #appending the decoded data to a list
df = pd.DataFrame (t1, columns = ['Text'])   #converting the list to data frame using pamdas
```

```python
df['Num of Words'] = df['Text'].apply(lambda y: len(str(y).split(" ")))
df
```

```python
df['Num of Char'] = df['Text'].str.len()
df
```

```python
#To find the average word length of the list
def awl(x):
    jai = x.split() #here the sentence is being splitted into parts and being assigned to variable jai
    if len(jai) != 0:  #here we are checking if the variable is empty or not
        return(sum(len(deep) for deep in jai)/len(jai))
    else:
        return None

df['Avg Word Length'] = df['Text'].apply(lambda z: awl(z))
df
```

```python
#To find the number of special characters present in the list

def spchar(x):
    red = 0
    for i in range(len(x)):
        if not((x[i].isalpha()) and (x[i].isdigit())):
            red = red + 1
    return red

df['Num of spec char'] = df['Text'].apply(lambda y: spchar(y))
df
```

```python
#To find the number of numerics present in the list

df['Num of num'] = df['Text'].apply(lambda x: len([x for x in x.split() if x.isdigit()])) #here we are considering it to be a seperate numeric if it is seperated with a space using the split function
df
```

```python
df['Num of upper case words'] = df['Text'].apply(lambda x: len([x for x in x.split() if x.isupper()])) #using split function we are finding the words with completely upper case
df
```

```python
df.to_csv('data.csv',index=True)
```

Question 4 (10 points): Write python code to rank the text based on text similarity. Based on the text data you used for question 2, design a query to match the most relevant docments. Please use the BERT model to represent both your query and the text data, then calculate the cosine similarity between the query and each text in your data. Rank the similary with descending order.

```python
# You code here (Please add comments in the code):
from transformers import BertTokenizer, BertModel
from sklearn.metrics.pairwise import cosine_similarity
import torch
import numpy as np

# Sample text data
documents = [
    "Natural language processing (NLP) is a field of artificial intelligence that focuses on the interaction between computers and humans through natural language.",
    "The ultimate objective of NLP is to enable computers to understand, interpret, and generate human language in a way that is both valuable and meaningful.",
    "Machine learning algorithms are often used in NLP to analyze and process large amounts of textual data.",
    "BERT (Bidirectional Encoder Representations from Transformers) is a pre-trained transformer model that has achieved state-of-the-art results in various NLP tasks."
]

# Your query
query = "What is the role of machine learning in natural language processing?"

# Load pre-trained BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Function to get BERT embeddings for a given text
def get_bert_embeddings(text):
    input_ids = tokenizer.encode(text, return_tensors='pt')
    with torch.no_grad():
        outputs = model(input_ids)
    embeddings = outputs.pooler_output.numpy()
    return embeddings

# Get BERT embeddings for the query and each document
query_embedding = get_bert_embeddings(query)
document_embeddings = [get_bert_embeddings(doc) for doc in documents]

# Calculate cosine similarity between the query and each document
similarities = [cosine_similarity(query_embedding, doc_embedding)[0][0] for doc_embedding in document_embeddings]

# Rank documents based on similarity in descending order
ranked_documents = sorted(zip(documents, similarities), key=lambda x: x[1], reverse=True)

# Display the ranked documents
print("Ranked Documents based on Cosine Similarity:")
for document, similarity in ranked_documents:
    print(f"Similarity: {similarity:.4f}\n{document}\n")



```
