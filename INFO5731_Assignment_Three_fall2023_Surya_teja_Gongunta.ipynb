# **Question 1: Understand N-gram**

(45 points). Write a python program to conduct N-gram analysis based on the dataset in your assignment two:

(1) Count the frequency of all the N-grams (N=3).

(2) Calculate the probabilities for all the bigrams in the dataset by using the fomular count(w2 w1) / count(w2). For example, count(really like) / count(really) = 1 / 3 = 0.33.

(3) Extract all the **noun phrases** and calculate the relative probabilities of each review in terms of other reviews (abstracts, or tweets) by using the fomular frequency (noun phrase) / max frequency (noun phrase) on the whole dataset. Print out the result in a table with column name the all the noun phrases and row name as all the 100 reviews (abstracts, or tweets).


```python
# Write your code here
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import string

# Function to clean text
def clean_text(text):
    # Lowercase the text
    text = text.lower()

    # Remove URLs
    text = re.sub(r'http\S+', '', text)

    # Remove non-alphabetic characters and digits
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Tokenize the text
    tokens = word_tokenize(text)

    # Lemmatize and remove stopwords
    lemmatizer = WordNetLemmatizer()
    stop_words = set(stopwords.words('english'))
    tokens = [lemmatizer.lemmatize(token) for token in tokens if token not in stop_words and len(token) > 1]

    # Join tokens back into a string
    cleaned_text = ' '.join(tokens)

    return cleaned_text

# Apply cleaning function to the 'reviews' column
df['cleaned_text'] = df['reviews'].apply(clean_text)

# Display the cleaned text
print(df[['reviews', 'cleaned_text']].head())
```

# Part 1: Count the frequency of all the N-grams (N=3)

```python
from collections import Counter
import nltk

text_data = df['cleaned_text']
# as given question is for N=3 am doing trigrams

# Tokenizing the text and generate trigrams
all_trigrams = []
for text in text_data:
    tokens = nltk.word_tokenize(text)
    trigrams = list(nltk.ngrams(tokens, 3))
    all_trigrams.extend(trigrams)

trigram_frequency = Counter(all_trigrams)

print(trigram_frequency.most_common(15))
```

# Part 2: Calculate the probabilities for all the bigrams

```python
# Assuming your text data is in a column named 'text'

# Tokenize the text and generate bigrams

# doing it for limited tweets as unable to get for all tweets

text_data = df['cleaned_text'][:500]

all_bigrams = []
for text in text_data:
    tokens = nltk.word_tokenize(text)
    bigrams = list(nltk.ngrams(tokens, 2))
    all_bigrams.extend(bigrams)

# Count the frequency of each word to calculate bigram probabilities
word_frequency = Counter()
for text in text_data:
    tokens = nltk.word_tokenize(text)
    word_frequency.update(tokens)

# Calculate bigram probabilities
bigram_probabilities = {}
for bigram in all_bigrams:
    w2, w1 = bigram[0], bigram[1]
    bigram_probabilities[bigram] = word_frequency[w2] / word_frequency[w1] if word_frequency[w1] != 0 else 0

# Display bigram probabilities
print(bigram_probabilities)
```

# Part 3: Extract noun phrases and calculate relative probabilities for each review. For this part, we'll use the NLTK library to extract noun phrases and calculate their relative probabilities for each review:

```python
from nltk import RegexpParser, pos_tag, word_tokenize
from collections import Counter

# Define a simple NP grammar
grammar = r'NP: {<DT>?<JJ>*<NN>}'
chunk_parser = RegexpParser(grammar)

# Extract noun phrases for each review and store in a list
first_500_reviews = df['cleaned_text'][:500]
all_noun_phrases = []

for review in first_500_reviews:
    tokens = word_tokenize(review)
    tagged = pos_tag(tokens)

    tree = chunk_parser.parse(tagged)

    noun_phrases = []
    for subtree in tree.subtrees():
        if subtree.label() == 'NP':
            noun_phrase = ' '.join(word for word, tag in subtree.leaves())
            noun_phrases.append(noun_phrase)

    all_noun_phrases.append(noun_phrases)

# Flatten the list of noun phrases
flat_noun_phrases = [noun_phrase for sublist in all_noun_phrases for noun_phrase in sublist]

# Calculate the frequency of each noun phrase
noun_phrase_frequency = Counter(flat_noun_phrases)

# Calculate relative probabilities for each review
relative_probabilities = {}
for i, noun_phrases in enumerate(all_noun_phrases):
    review_probabilities = {noun_phrase: noun_phrase_frequency[noun_phrase] / max(noun_phrase_frequency.values()) for noun_phrase in noun_phrases}
    relative_probabilities[f'Review {i+1}'] = review_probabilities

# Display the relative probabilities as a table
relative_probabilities_df = pd.DataFrame(relative_probabilities).fillna(0)
print(relative_probabilities_df)
```

# **Question 2: Undersand TF-IDF and Document representation**

(20 points). Starting from the documents (all the reviews, or abstracts, or tweets) collected for assignment two, write a python program:

(1) To build the **documents-terms weights (tf*idf) matrix bold text**.

(2) To rank the documents with respect to query (design a query by yourself, for example, "An Outstanding movie with a haunting performance and best character development") by using **cosine similarity**.


# (1) To build the **documents-terms weights (tf*idf) matrix bold text**.

```python
# Write your code here

from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize the TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the text data to TF-IDF matrix
tfidf_matrix = tfidf_vectorizer.fit_transform(df['cleaned_text'])

# Get the feature names (terms)
terms = tfidf_vectorizer.get_feature_names_out()

# Create a DataFrame for TF-IDF matrix
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=terms)

# Display the TF-IDF matrix
print(tfidf_df)
```

# (2) To rank the documents with respect to query (design a query by yourself, for example, "An Outstanding movie with a haunting performance and best character development") by using **cosine similarity**.

```python
from sklearn.metrics.pairwise import cosine_similarity

# Define your query
query = "Surya is trying hard to understand natural language processing and Machine Learning"

# Transform the query into TF-IDF vector
query_vector = tfidf_vectorizer.transform([query])

# Calculate cosine similarity between query and all documents
cosine_similarities = cosine_similarity(query_vector, tfidf_matrix)

# Get document indices sorted by similarity
sorted_indices = cosine_similarities.argsort()[0][::-1]

# Display documents ranked by similarity to the query
for idx in sorted_indices:
    print(f"Document {idx+1}: Similarity = {cosine_similarities[0][idx]}")
    print(df['cleaned_text'][idx])
    print("-" * 50)
```

# (1) Train a 300-dimension word embedding (it can be word2vec, glove, ulmfit, bert, or others).

```python
# Write your code here

from nltk.tokenize import word_tokenize
from gensim.models import Word2Vec

text_data=df['reviews'][:500]

# Tokenize your text data into words
tokenized_data = [word_tokenize(sentence.lower()) for sentence in text_data]

# Train Word2Vec model
model = Word2Vec(sentences=tokenized_data, vector_size=300, window=5, min_count=1, workers=4)
```

(2) Visualize the word embedding model you created.

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Extract word vectors
word_vectors = model.wv

# Reduce dimensions to visualize in 2D
words = list(word_vectors.index_to_key)
vectors = model.wv[words]
pca = PCA(n_components=2)
vectors_2d = pca.fit_transform(vectors)

# Plot word vectors in 2D
plt.figure(figsize=(10, 8))
plt.scatter(vectors_2d[:, 0], vectors_2d[:, 1], edgecolors='k', c='r')
for word, (x, y) in zip(words, vectors_2d):
    plt.text(x, y, word)
plt.title('Word Embedding Visualization')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.grid(True)
plt.show()
```
